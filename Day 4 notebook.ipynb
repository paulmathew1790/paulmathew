```python
import numpy as np

a = np.array([[1, 2],
              [4, 5]])

b = np.array([[7, 8],
              [10, 11]])

c = a @ b
print(c)
print("______________")
d = a+b
print(d)
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

elementwise = A * B
matrix_product = A @ B

print("Element-wise :\n", elementwise)
print("Matrix (dot) product:\n", matrix_product)
```

```python
A = np.array([
    [1,2],
    [3,4]
])

print(A.T)
```

```python
A = np.array([
    [1,2],
    [3,4]
])

print(np.linalg.inv(A))
```

```python
A = np.array([
    [2,1],
    [3,4]
])

print(np.linalg.det(A))
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print("Addition")
print(A+B)

print("\nSubtraction")
print(A-B)

print("\nElement-wise Multiplication")
print(A*B)
print("\n dot Multiplication")
print(A @B)
```

```python
tensor = np.array([
    [[1,2],[3,4]],
    [[5,6],[7,8]]
])

print(tensor)

print("Dimensions:", tensor.ndim)
print("Shape:", tensor.shape)
```

```python
A = np.array([
    [4,2],
    [1,3]
])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues")
print(eigenvalues)

print("\nEigenvectors")
print(eigenvectors)
```

```python
import numpy as np

A = np.array([
    [2,0],
    [0,3]
])

values, vectors = np.linalg.eig(A)

print(values)
print("_____")
print(vectors)
```
