```python
import numpy as np

A = np.array([[2, 0, 0],
              [0, 3, 4],
              [0, 4, 9]], dtype=float)

eigenvalues, P = np.linalg.eig(A)
D = np.diag(eigenvalues)
P_inv = np.linalg.inv(P)

A_reconstructed = P @ D @ P_inv

print("Eigenvalues (diagonal of D):", eigenvalues)
print("Reconstructed A:\n", A_reconstructed.real)
print("Matches original?", np.allclose(A, A_reconstructed.real))
```

```python
import numpy as np

np.random.seed(8)
n_tosses = 10000
tosses = np.random.choice(["H", "T"], size=n_tosses)

n_heads = np.sum(tosses == "H")
empirical_p_heads = n_heads / n_tosses

print(f"Heads: {n_heads}/{n_tosses}")
print(f"Empirical P(Heads) = {empirical_p_heads:.4f}")
print("Theoretical P(Heads) = 0.5")
```

```python
import numpy as np

P_disease = 0.01
P_no_disease = 1 - P_disease
P_pos_given_disease = 0.95        # sensitivity
P_neg_given_no_disease = 0.90     # specificity
P_pos_given_no_disease = 1 - P_neg_given_no_disease   # false positive rate

# Bayes' theorem:
# P(disease | positive) = P(positive | disease) * P(disease) / P(positive)
P_positive = (P_pos_given_disease * P_disease) + (P_pos_given_no_disease * P_no_disease)
P_disease_given_positive = (P_pos_given_disease * P_disease) / P_positive

print(f"P(positive test) = {P_positive:.4f}")
print(f"P(disease | positive test) = {P_disease_given_positive:.4f}")
```

```python
import numpy as np

data = np.array([12, 15, 14, 10, 18, 20, 9, 16])

variance = np.var(data)
std_dev = np.std(data)

print(f"Variance: {variance:.4f}")
print(f"Standard deviation: {std_dev:.4f}")
print(f"sqrt(variance) == std_dev? {np.isclose(np.sqrt(variance), std_dev)}")
```

```python
import numpy as np

sample_data = np.array([23, 19, 27, 21, 25])

pop_var = np.var(sample_data, ddof=0)
sample_var = np.var(sample_data, ddof=1)

print(f"Population variance (ddof=0): {pop_var:.4f}")
print(f"Sample variance (ddof=1):     {sample_var:.4f}")

n = len(sample_data)
manual_sample_var = np.sum((sample_data - sample_data.mean())**2) / (n - 1)
print(f"Manual sample variance (n-1): {manual_sample_var:.4f}")
```

```python
import numpy as np

data = np.array([50, 60, 70, 80, 90, 100, 40, 65])

mean_val = data.mean()
std_val = data.std()

z_scores = (data - mean_val) / std_val

print("Z-scores:", np.round(z_scores, 3))
print(f"New mean: {z_scores.mean():.6f}")
print(f"New std:  {z_scores.std():.6f}")
```

```python
import numpy as np

x = np.array([2, 4, 6, 8, 10, 12])
y = np.array([1, 3, 5, 6, 9, 15])

x_mean, y_mean = x.mean(), y.mean()
covariance = np.sum((x - x_mean) * (y - y_mean)) / len(x)
std_x = np.std(x)
std_y = np.std(y)

manual_r = covariance / (std_x * std_y)
builtin_r = np.corrcoef(x, y)[0, 1]

print(f"Manual Pearson r: {manual_r:.4f}")
print(f"np.corrcoef r:    {builtin_r:.4f}")
print("Match?", np.isclose(manual_r, builtin_r))
```

```python
import numpy as np

def f(x):
    return x ** 3

def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

def exact_derivative(x):
    return 3 * x ** 2      # d/dx[x^3] = 3x^2

x_points = np.array([-2, -1, 0, 1, 2, 3])
numeric = numerical_derivative(f, x_points)
exact = exact_derivative(x_points)

print("x:", x_points)
print("Numerical derivative:", np.round(numeric, 5))
print("Exact derivative:    ", exact)
```
