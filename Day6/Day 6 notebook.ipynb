## 3. Broadcasting

Broadcasting lets NumPy perform operations on arrays of different shapes.

```python
a = np.array([1, 2, 3])
b = np.array([[1], [2], [3]])

# Adding a (1,3) array to a (3,1) array -> broadcasts to (3,3)
result = a + b
print(result)

# Scalar broadcasting
scaled = arr_1d * 2
print('Scaled array:', scaled)
```

```python
import numpy as np
import pandas as pd

# Sample Dataset
data = {
    "Name": ["John", "Alice", "Bob", "John", None],
    "Age": [23, np.nan, 28, 23, 30],
    "Salary": [50000, 60000, np.nan, 50000, 70000],
    "Department": ["IT", "HR", "IT", "IT", "Sales"]
}

df = pd.DataFrame(data)

print("Original Dataset")
print(df)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Fill missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
df["Name"] = df["Name"].fillna("Unknown")

# Feature Engineering
df["Salary_Lakh"] = df["Salary"] / 100000
df["Bonus"] = df["Salary"] * 0.10
df["Age_Group"] = df["Age"].apply(lambda x: "Young" if x < 30 else "Adult")

print("\nCleaned Dataset")
print(df)

print("\nAverage Salary by Department")
print(df.groupby("Department")["Salary"].mean())

# Save cleaned dataset
df.to_csv("cleaned_dataset.csv", index=False)
```

```python
print(df[df["Age"] > 30])
print(df[df["Salary"] > 60000])
print(df[df["Department"] == "IT"])
print(df[(df["Age"] > 30) & (df["Salary"] > 60000)])
```

```python
print(df.groupby("Department")["Salary"].mean())
print(df.groupby("Department")["Salary"].max())
print(df.groupby("Department")["Salary"].agg(["mean", "max", "min", "count"]))
print(df.groupby("Department")["Employee_ID"].count())
```
