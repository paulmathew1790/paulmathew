```python
# 1) Classes, Objects, and Constructors
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def is_passing(self):
        return self.grade >= 60

student = Student("Ada", 88)
print("Student:", student.name, student.is_passing())
```

```python
# 2) Inheritance
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, I am {self.name}."

class Teacher(Person):
    def __init__(self, name, subject):
        super().__init__(name)
        self.subject = subject

    def greet(self):
        return f"Hello, I am {self.name} and I teach {self.subject}."

teacher = Teacher("Mr. Smith", "Python")
print(teacher.greet())
```

```python
# 3) Polymorphism
class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

for animal in [Dog(), Cat()]:
    print(animal.speak())
```

```python
# 4) Encapsulation and Abstraction
from abc import ABC, abstractmethod

class BankAccount(ABC):
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

    @abstractmethod
    def account_type(self):
        pass

class SavingsAccount(BankAccount):
    def account_type(self):
        return "Savings"

account = SavingsAccount()
account.deposit(100)
account.withdraw(20)
print("Balance:", account.get_balance())
print("Account type:", account.account_type())
```

```python
def square(n):
    for i in range(1, n + 1):
        yield i ** 2


for value in square(5):
    print(value)
```

```python
def decorator(func):

    def wrapper():
        print("Before Function")

        func()

        print("After Function")

    return wrapper


@decorator
def greet():
    print("Hello")


greet()
```

```python
import time

def timer(func):

    def wrapper():
        start = time.time()

        func()

        end = time.time()

        print("Execution Time:", end - start)

    return wrapper


@timer
def task():
    for i in range(1000000):
        pass


task()
```

```python
import random
import requests


class Holding:
    def __init__(self, symbol, quantity, buy_price):
        self.symbol = symbol
        self.quantity = quantity
        self.buy_price = buy_price

    def investment(self):
        return self.quantity * self.buy_price


class Portfolio:
    def __init__(self):
        self.holdings = []

    def add_holding(self, holding):
        self.holdings.append(holding)

    def analyze(self, live_prices):
        total_investment = 0
        total_value = 0

        print("\n" + "=" * 70)
        print("              CRYPTO PORTFOLIO ANALYSIS")
        print("=" * 70)

        for holding in self.holdings:

            investment = holding.investment()

            current_price = live_prices[holding.symbol]

            current_value = holding.quantity * current_price

            profit = current_value - investment

            total_investment += investment
            total_value += current_value

            print(f"""
Coin           : {holding.symbol}
Quantity       : {holding.quantity}

Buy Price      : ${holding.buy_price:,.2f}
Current Price  : ${current_price:,.2f}

Investment     : ${investment:,.2f}
Current Value  : ${current_value:,.2f}
Profit/Loss    : ${profit:,.2f}
""")
            print("-" * 70)

        print("\nSUMMARY")
        print("=" * 70)
        print(f"Total Investment : ${total_investment:,.2f}")
        print(f"Current Value    : ${total_value:,.2f}")
        print(f"Net Profit/Loss  : ${total_value-total_investment:,.2f}")

        if total_value > total_investment:
            print("\nPortfolio Status : PROFIT 📈")
        elif total_value < total_investment:
            print("\nPortfolio Status : LOSS 📉")
        else:
            print("\nPortfolio Status : BREAK EVEN")


def get_live_prices():

    url = "https://api.coingecko.com/api/v3/simple/price"

    params = {
        "ids": "bitcoin,ethereum,solana,cardano",
        "vs_currencies": "usd"
    }

    response = requests.get(url, params=params, timeout=10)

    data = response.json()

    return {
        "BTC": data["bitcoin"]["usd"],
        "ETH": data["ethereum"]["usd"],
        "SOL": data["solana"]["usd"],
        "ADA": data["cardano"]["usd"]
    }


def generate_random_portfolio():

    portfolio = Portfolio()

    portfolio.add_holding(
        Holding(
            "BTC",
            round(random.uniform(0.1, 2), 2),
            random.randint(30000, 70000),
        )
    )

    portfolio.add_holding(
        Holding(
            "ETH",
            round(random.uniform(1, 10), 2),
            random.randint(1500, 4000),
        )
    )

    portfolio.add_holding(
        Holding(
            "SOL",
            round(random.uniform(5, 50), 2),
            random.randint(80, 220),
        )
    )

    portfolio.add_holding(
        Holding(
            "ADA",
            round(random.uniform(50, 500), 2),
            round(random.uniform(0.3, 1.2), 2),
        )
    )

    return portfolio


def main():

    print("Generating Random Portfolio...")

    portfolio = generate_random_portfolio()

    print("Fetching Live Market Prices...")

    try:
        prices = get_live_prices()

        print("Live prices fetched successfully.\n")

    except Exception:

        print("Could not fetch live prices.")
        print("Using backup prices.\n")

        prices = {
            "BTC": 60000,
            "ETH": 3000,
            "SOL": 150,
            "ADA": 0.50,
        }

    portfolio.analyze(prices)


if __name__ == "__main__":
    main()
```

```python
import random
import requests
from abc import ABC, abstractmethod



class CryptoAsset(ABC):

    def __init__(self, symbol, quantity, buy_price):
        self._symbol = symbol
        self._quantity = quantity
        self._buy_price = buy_price

    @property
    def symbol(self):
        return self._symbol

    @property
    def quantity(self):
        return self._quantity

    @property
    def buy_price(self):
        return self._buy_price

    @abstractmethod
    def investment(self):
        """Must be implemented by every asset"""
        pass




class Holding(CryptoAsset):

    def __init__(self, symbol, quantity, buy_price):
        super().__init__(symbol, quantity, buy_price)

    def investment(self):
        return self.quantity * self.buy_price



class Portfolio:

    def __init__(self):
        self.holdings = []

    def add_holding(self, holding):
        self.holdings.append(holding)

    def total_investment(self):
        return sum(asset.investment() for asset in self.holdings)

    def total_value(self, prices):

        total = 0

        for asset in self.holdings:
            total += asset.quantity * prices[asset.symbol]

        return total

    def profit(self, prices):
        return self.total_value(prices) - self.total_investment()

    def analyze(self, live_prices):

        print("\n" + "=" * 70)
        print("             CRYPTO PORTFOLIO ANALYSIS")
        print("=" * 70)

        for asset in self.holdings:

            investment = asset.investment()

            current_price = live_prices[asset.symbol]

            current_value = asset.quantity * current_price

            profit = current_value - investment

            print(f"""
Coin           : {asset.symbol}
Quantity       : {asset.quantity}

Buy Price      : ${asset.buy_price:,.2f}
Current Price  : ${current_price:,.2f}

Investment     : ${investment:,.2f}
Current Value  : ${current_value:,.2f}
Profit/Loss    : ${profit:,.2f}
""")

            print("-" * 70)

        total_investment = self.total_investment()
        total_value = self.total_value(live_prices)
        net_profit = self.profit(live_prices)

        print("\nSUMMARY")
        print("=" * 70)

        print(f"Total Investment : ${total_investment:,.2f}")
        print(f"Current Value    : ${total_value:,.2f}")
        print(f"Net Profit/Loss  : ${net_profit:,.2f}")

        if net_profit > 0:
            print("\nPortfolio Status : PROFIT 📈")

        elif net_profit < 0:
            print("\nPortfolio Status : LOSS 📉")

        else:
            print("\nPortfolio Status : BREAK EVEN")


# =====================================================
# REST API
# =====================================================

def get_live_prices():

    url = "https://api.coingecko.com/api/v3/simple/price"

    params = {
        "ids": "bitcoin,ethereum,solana,cardano",
        "vs_currencies": "usd"
    }

    response = requests.get(url, params=params, timeout=10)

    response.raise_for_status()

    data = response.json()

    return {
        "BTC": data["bitcoin"]["usd"],
        "ETH": data["ethereum"]["usd"],
        "SOL": data["solana"]["usd"],
        "ADA": data["cardano"]["usd"]
    }


# =====================================================
# Generate Random Portfolio
# =====================================================

def generate_random_portfolio():

    portfolio = Portfolio()

    portfolio.add_holding(
        Holding(
            "BTC",
            round(random.uniform(0.1, 2), 2),
            random.randint(30000, 70000)
        )
    )

    portfolio.add_holding(
        Holding(
            "ETH",
            round(random.uniform(1, 10), 2),
            random.randint(1500, 4000)
        )
    )

    portfolio.add_holding(
        Holding(
            "SOL",
            round(random.uniform(5, 50), 2),
            random.randint(80, 220)
        )
    )

    portfolio.add_holding(
        Holding(
            "ADA",
            round(random.uniform(50, 500), 2),
            round(random.uniform(0.30, 1.20), 2)
        )
    )

    return portfolio




    def main():

        print("Generating Random Portfolio...")

        portfolio = generate_random_portfolio()

        print("Fetching Live Market Prices...")

        try:

            prices = get_live_prices()

            print("Live Prices Retrieved Successfully.\n")

        except Exception as e:

            print("API Failed")
            print(e)

            print("\nUsing Backup Prices...\n")

            prices = {
                "BTC": 60000,
                "ETH": 3000,
                "SOL": 150,
                "ADA": 0.50
            }

        portfolio.analyze(prices)




    if __name__ == "__main__":
        main()
```
