```python
def linear_search(arr, target):
    # Check every element in the list
    for i in range(len(arr)):
        if arr[i] == target:
            return i  # Return index if found

    return -1  # Return -1 if not found


numbers = [10, 20, 30, 40, 50]

target = 30

result = linear_search(numbers, target)
print(result)
```

```python
def binary(arr,target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
r = binary([10, 20, 30, 40, 50], 50)
print(r)
```

```python
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):

        for j in range(0, n - i - 1):

            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr


numbers = [64, 34, 25, 12, 22]

print("Sorted List:", bubble_sort(numbers))
```

```python
def insertion_sort(arr):

    for i in range(1, len(arr)):

        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr


numbers = [12, 11, 13, 5, 6]

print(insertion_sort(numbers))
```

```python
def fact(n):
  
    # BASE CONDITION
    if n == 0:
        return 1
    return n * fact(n - 1)

print("Factorial of 5 : ", fact(5))
```

```python
def quick_sort(arr):

    if len(arr) <= 1:
        return arr

    pivot = arr[0]

    left = []
    right = []

    for num in arr[1:]:

        if num <= pivot:
            left.append(num)
        else:
            right.append(num)

    return quick_sort(left) + [pivot] + quick_sort(right)


numbers = [50, 23, 9, 18, 61, 32]

print(quick_sort(numbers))
```

```python
def merge_sort(arr):

    if len(arr) > 1:

        mid = len(arr) // 2

        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

        while i < len(left) and j < len(right):

            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1

            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1

    return arr


numbers = [38, 27, 43, 3, 9, 82, 10]

print(merge_sort(numbers))
```

```python
def selection_sort(arr):
    n = len(arr)

    for i in range(n):

        min_index = i

        for j in range(i + 1, n):

            if arr[j] < arr[min_index]:
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr


numbers = [29, 10, 14, 37, 13]

print(selection_sort(numbers))
```
