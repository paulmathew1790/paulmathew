```python
# Problem 1: Add two numbers
num1 = 10
num2 = 20
print("Sum:", num1 + num2)
```

```python
# Problem 2: Find the area of a circle
radius = 7
area = 3.14 * radius * radius
print("Area:", area)
```

```python
# Problem 3: Swap two variables
a = 5
b = 9
a, b = b, a
print("a =", a)
print("b =", b)
```

```python
# Problem 4: Check even or odd
number = 13
if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Problem 5: Check leap year
year = 2024
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap year")
else:
    print("Not a leap year")
```

```python
# Problem 6: Find the largest of three numbers
x, y, z = 12, 45, 27
print("Largest:", max(x, y, z))
```

```python
# Problem 7: Print multiplication table
n = 5
for i in range(1, 11):
    print(f"{n} x {i} = {n * i}")
```

```python
# Problem 8: Sum of the first n natural numbers
n = 10
print("Sum:", n * (n + 1) // 2)
```

```python
# Problem 9: Reverse a string
word = "python"
print(word[::-1])
```

```python
# Problem 10: Count vowels in a string
text = "programming"
vowels = "aeiou"
count = sum(1 for ch in text if ch in vowels)
print("Vowel count:", count)
```

```python
# Problem 11: Check palindrome
word = "level"
print(word == word[::-1])
```

```python
# Problem 12: Find factorial
n = 5
factorial = 1
for i in range(1, n + 1):
    factorial *= i
print("Factorial:", factorial)
```

```python
# Problem 13: Print the first n Fibonacci numbers
n = 7
fib = [0, 1]
for _ in range(2, n):
    fib.append(fib[-1] + fib[-2])
print(fib)
```

```python
# Problem 14: Generate prime numbers up to 20
primes = []
for num in range(2, 21):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)
print(primes)
```

```python
# Problem 15: Merge two lists and remove duplicates
list1 = [1, 2, 2, 3]
list2 = [3, 4, 5, 5]
merged = list(dict.fromkeys(list1 + list2))
print(merged)
```

```python
# Problem 16: Count character frequency
text = "banana"
freq = {}
for ch in text:
    freq[ch] = freq.get(ch, 0) + 1
print(freq)
```

```python
# Problem 17: Find the second largest number
nums = [10, 20, 4, 45, 99]
nums = sorted(nums)
print("Second largest:", nums[-2])
```

```python
# Problem 18: Calculate average of numbers
values = [10, 20, 30, 40]
print("Average:", sum(values) / len(values))
```

```python
# Problem 19: Create a function to calculate power
def power(base, exponent):
    return base ** exponent

print(power(2, 5))
```

```python
# Problem 20: Use a lambda function to square numbers
square = lambda x: x * x
print(square(8))
```

```python
# Problem 21: List comprehension for even numbers
evens = [n for n in range(1, 11) if n % 2 == 0]
print(evens)
```

```python
# Problem 22: Set comprehension for unique squares
squares = {n * n for n in range(1, 6)}
print(squares)
```

```python
# Problem 23: Dictionary comprehension from a list
words = ["apple", "banana", "cherry"]
word_lengths = {word: len(word) for word in words}
print(word_lengths)
```

```python
# Problem 24: Handle division by zero safely
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```
