## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


print('Matplotlib ready')
```

## 2. Sample Dataset

We'll create a small sample dataset to use across all visualizations.

```python
np.random.seed(42)

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Henry'],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT'],
    'Age': [25, 30, 35, 22, 28, 40, 33, 27],
    'Salary': [50000, 65000, 60000, 48000, 72000, 55000, 58000, 67000],
    'Experience': [2, 6, 9, 1, 5, 15, 8, 4],
}

df = pd.DataFrame(data)
df
```

## 3. Basic Charts

### 3.1 Bar Chart - Salary by Employee

```python
plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Salary'], color='steelblue')
plt.title('Salary by Employee')
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 3.2 Line Chart - Experience Trend

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Name'], df['Experience'], marker='o', color='darkorange')
plt.title('Years of Experience by Employee')
plt.xlabel('Employee')
plt.ylabel('Experience (years)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 3.3 Pie Chart - Department Distribution

```python
dept_counts = df['Department'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%', startangle=90,
        colors=['#4C72B0', '#DD8452', '#55A868'])
plt.title('Employee Distribution by Department')
plt.axis('equal')
plt.show()
```

## 4. Histograms

Histograms show the distribution of a single numeric variable.

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Salary'], bins=5, color='mediumseagreen', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Larger random dataset to better illustrate distribution shape
random_ages = np.random.normal(loc=30, scale=6, size=500)

plt.figure(figsize=(8, 5))
plt.hist(random_ages, bins=20, color='cornflowerblue', edgecolor='black')
plt.title('Simulated Age Distribution (n=500)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## 5. Scatter Plots

Scatter plots reveal relationships between two numeric variables.

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Experience'], df['Salary'], color='crimson', s=80)
plt.title('Experience vs Salary')
plt.xlabel('Experience (years)')
plt.ylabel('Salary')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot colored by department
colors = {'HR': 'tab:blue', 'IT': 'tab:orange', 'Finance': 'tab:green'}

plt.figure(figsize=(8, 5))
for dept, group in df.groupby('Department'):
    plt.scatter(group['Age'], group['Salary'], label=dept, color=colors[dept], s=80)

plt.title('Age vs Salary by Department')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend(title='Department')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

## 6. Correlation Heatmap (Concept)

A correlation heatmap visualizes how strongly pairs of numeric variables are related, using color intensity to represent the correlation coefficient (ranging from -1 to 1). We can build one using only Matplotlib (no seaborn required).

```python
numeric_df = df[['Age', 'Salary', 'Experience']]
corr_matrix = numeric_df.corr()
corr_matrix
```

```python
fig, ax = plt.subplots(figsize=(6, 5))
cax = ax.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)

# Axis labels
ax.set_xticks(range(len(corr_matrix.columns)))
ax.set_yticks(range(len(corr_matrix.columns)))
ax.set_xticklabels(corr_matrix.columns)
ax.set_yticklabels(corr_matrix.columns)

# Annotate each cell with its correlation value
for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        ax.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}', ha='center', va='center', color='black')

ax.set_title('Correlation Heatmap (Age, Salary, Experience)')
fig.colorbar(cax, label='Correlation coefficient')
plt.tight_layout()
plt.show()
```
